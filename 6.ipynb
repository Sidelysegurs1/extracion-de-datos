```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Cargar datos de acciones de GameStop
gamestop_stock = yf.download('GME', start='2010-01-01', end='2022-02-26')
```

```python
# Cargar datos de ingresos de GameStop
gamestop_incomes = pd.read_html('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html')[0]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(gamestop_stock.index, gamestop_stock['Close'], label='GameStop Stock Price')
plt.xlabel('Fecha')
plt.ylabel('Precio de la accin (USD)')
plt.title('Evolucin de las acciones de GameStop')
plt.legend()
plt.show()
```
